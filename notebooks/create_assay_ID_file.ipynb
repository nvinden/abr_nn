```python
import pandas as pd

# Input File
amr_data = "/home/nvinden/School/CIS4020/kkinds_ABR/data/full_data_NY/AMR_Data.csv"
save_name = "/home/nvinden/School/CIS4020/kkinds_ABR/data/full_data_NY/assay_keys.csv"
```

```python
df = pd.read_csv(amr_data)
```

```python
# Get all unique assay names
assay_names = df['assay_name'].unique()

# Create a DataFrame with these unique names in both columns
# You need to ensure the data is structured correctly to create the two columns
assay_df = pd.DataFrame({'assay_name': assay_names, 'assay_cat': assay_names})

# Replace empty strings with "UNKNOWN"
assay_df = assay_df.replace(r'^\s*$', 'UNKNOWN', regex=True)

# Save the dataframe
assay_df.to_csv(save_name, index=False)

assay_df
```
